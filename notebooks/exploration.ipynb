# Exploration du dataset Uber
Ce notebook explore les premières lignes du dataset et propose quelques visualisations simples.

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Uber Exploration").getOrCreate()
df = spark.read.csv("../data/raw/UberDataset.csv", header=True, inferSchema=True)
df.show(5)
```

```python
# Statistiques descriptives
df.describe().show()
```

```python
# Nombre de trajets par catégorie
df.groupBy("CATEGORY").count().show()
```
